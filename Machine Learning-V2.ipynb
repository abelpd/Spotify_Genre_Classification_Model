## ------------------------------------------One vs Rest -------------------------------------------

```python
#import libraries
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
#create models to apply to one vs rest algorithm
deciders = {'Naive Bayes':MultinomialNB(), 'LogRegression':LogisticRegression(), 'DecisionTree': DecisionTreeClassifier()}
```

```python
#convert loudness to positive values (model will not work with negative features values)
X_train.loudness = abs(X_train.loudness)
```

```python
# iterate through the deciders
# track the best OvR model and input to the ModelAccuracy dictionary
best_X_test = 0
best_X_train = 0
decid = ""
for key,value in deciders.items(): 
    clf = OneVsRestClassifier(value)
    clf.fit(X_train,y_train)
    clfpred = clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, clfpred)
    train_accuracy = accuracy_score(y_train, clf.predict(X_train))
    if test_accuracy > best_X_test:
        best_X_test = test_accuracy
        best_X_train = train_accuracy
        decid = key
    print (key + ': test - ' + str(test_accuracy) + ' | train - ' + str(train_accuracy))
    
ModelAccuarcy['OvR_' + decid] = {'y_test': best_X_test, 'y_train': best_X_train}
```
