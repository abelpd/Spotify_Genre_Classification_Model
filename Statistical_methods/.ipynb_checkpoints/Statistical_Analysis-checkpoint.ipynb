### Statistical Analysis
### Questions to answer:
### <span style="color:blue"><u>1) From the basic EDA, we see that featuring an artist does lead to higher stream numbers on average. However, is this observed difference statistically significant?</u></span><br>
<b>After performing 10,000 bootstrap samples on adjusted means, we observe a pvalue that is extremely small. We can include that yes, there is a network effect in which featuring another artist will on average translate to higher stream numbers.</b>

### <span style="color:blue"><u>2a) Looking at the features to describe genre, is there any relationship between any of the features? If so, will THE correlation between two features impede a ML model?</u></span><br>
<b>Creating a correltation matrix of the features, we see a few correlations between the features.  
Most notably, we see negative correlation between accousticness and energy and a positive corrleation between loudness and energy.</b>

### <span style="color:blue"><u>2b) Will correlation between two features impede a ML model?</u></span><br>  
##### To determine if  correlation between features will impede the ML model, a regression was run to see if there is a significant R^2 between the two features. For example, if we conclude that energy is a good predictor for accousticness, we can be reasonably sure that one of the features can be excluded from the final model.

##### <u>Accousticness & Energy </u>
##### WIth an R^2 of 0.37, we can state that energy predicts roughly 37% of acousticness in a song. This is not significant enough to exclude either variable from our ML model.

##### <u>Loudness & Energy </u>
##### IMPORTANT TO NOTE: Loudness in this dataset is measured in decibles from a baseline 0. Furthermore, decibles are usually measured as a logarithm (base 10) to encapsulate the wide range of numeric values that decibles can take on. The most accurate manner to determine correlation would be to take the exponent of loudness and scale using standard scaling. For my purposes, I skipped the exponent part. This does introduce a small amount of inaccuracy in the linear model, however, it is likely not significant enough to bias the end result.

##### Loudness & Energy have an R^2 of 39%. Once again this is not significant enough to exclude one of the features in the final model.

### <span style="color:blue"><u>3) Is there a meaningful difference between subgenres in terms of features? Put another way, is there meaningful difference between say 'pop' music and 'dance pop' music?</u></span><br>  
##### Taking bootstrap samples of each feature and a similar genres (ex. pop & dance_pop), we see that only a handful of features are significantly different.  
##### Once major concern is that keeping each subgenre separate might induce a high amount of error in the model. This is due to the sparcity induced by having few features that are statistically significant between  subgenres. A possible response is to group similar genres.
##### Both the original dataset and a condensed version will be used to determine the optimal data for the model.

## 1) From the basic EDA, we see that featuring an artist does lead to higher stream numbers on average. However, is this observed difference statistically significant?

```python
#Looking through a few of the csv files, Spotify signifies that an artist is featuring someone when the title
#of the song has the keywords "feat" or "with".

#regex to find songs that feature another artist
#df['is_featured'] = df['is_featured'].replace(to_replace=True, value=1, inplace=True, regex=False)
df['is_featured'] = df['Track Name'].str.contains(r'([\(]+feat|[\(]+with)',regex=True)
#df['is_featured'] = df['is_featured'].apply(lambda x: 'featured' if x == 'NaN' else 'Not Featured')
df.head(5)
```

```python
#determine if a featured song generally has more streams
is_featured_mean = df.groupby('is_featured')['Streams'].mean()
is_featured_mean.index = ['non-featured', 'featured']
is_featured_std = df.groupby('is_featured')['Streams'].std()
is_featured_std.index = ['non-featured', 'featured']

print (is_featured_mean)
print (is_featured_std)
```

```python
#To determine if the means are significantly different, a bootstrap sampling of the means was performed

#loading the songs into numpy arrays
featured_songs = np.array(df['Streams'][df['is_featured']==1])
not_featured_songs = np.array(df['Streams'][df['is_featured']==0])

#calculate the difference in means
difference_in_mean = np.mean(featured_songs) - np.mean(not_featured_songs)

#compute mean stream of both samples
mean_streams = np.mean(np.append(featured_songs, not_featured_songs))
print (mean_streams)

#adjusting the two samples to have the same mean
featured_songs = featured_songs - np.mean(featured_songs) + mean_streams
not_featured_songs = not_featured_songs - np.mean(not_featured_songs) + mean_streams

print (np.mean(featured_songs))
print (np.mean(not_featured_songs))
print ('The difference between the sample means is %s' %difference_in_mean)
```

```python
#generate replicates for each sample
featured_replicates = np.array([np.random.choice(featured_songs, len(not_featured_songs), replace=True).mean() for x in range(10000)])
not_featured_replicates = np.array([np.random.choice(not_featured_songs, len(not_featured_songs), replace=True).mean() for x in range(10000)])
```

```python
replicate_differences = featured_replicates - not_featured_replicates
pvalue = np.sum(replicate_differences >= difference_in_mean) / len(replicate_differences)
print (pvalue)
```

## 3) Is there a meaningful difference between subgenres in terms of features? Put another way, can the features describe the difference between say 'pop' music and 'dance pop' music?

###### The goal is to compare the features between subgenres to see if they can be grouped prior to running through the final model. Is discern if there is a difference between the features of common genres, a bootstrap is performed. The procedure will be run on the following grouped genres:

* pop / dance_pop / post_teen_pop
* pop_rap / hip_hop / rap
* rock / modern_rock
* edm / topical_house

###### **pop_rap was included in the rap catagory due to its simialrity to rap/hip hop and dissimilarity to pop (See EDA section for specifics)

```python
y['row'] = np.sum(y,axis=1)
```

```python
#create genre subsets
pop = X[(y['pop'] == 1) & (y['row']==1)]
dance_pop = X[(y['dance_pop'] == 1) & (y['row']==1)]
pop_rap = X[(y['pop_rap'] == 1) & (y['row']==1)]
rock = X[(y['rock'] == 1) & (y['row']==1)]
tropical_house = X[(y['tropical_house'] == 1) & (y['row']==1)]
hip_hop = X[(y['hip_hop'] == 1) & (y['row']==1)]
modern_rock = X[(y['modern_rock'] == 1) & (y['row']==1)]
post_teen_pop = X[(y['post_teen_pop'] == 1) & (y['row']==1)]
edm = X[(y['edm'] == 1) & (y['row']==1)]
rap = X[(y['rap'] == 1) & (y['row']==1)]
```

```python
def bootstrap_features(g1, g2):
    #To determine if the means are significantly different, a bootstrap sampling of the means was performed

    genre_diff = []
    feature_iter = g1.columns

    
    #loading features into numpy arrays
    for genre in g1:
        g1f = np.array(g1[genre])
        g2f = np.array(g2[genre])

        #determine minimum number of samples to extract
        min_rep = min(len(g1),len(g2))
        
        #calculate the difference in means
        difference_in_mean = np.mean(g1f) - np.mean(g2f)

        #compute mean stream of both samples
        mean_ = np.mean(np.append(g1f, g2f))

        #adjusting the two samples to have the same mean
        g1f = g1f - np.mean(g1f) + mean_
        g2f = g2f - np.mean(g2f) + mean_

        #generate replicates for each sample
        g1_replicates = np.array([np.random.choice(g1f, min_rep, replace=True).mean() for x in range(10000)])
        g2_replicates = np.array([np.random.choice(g2f, min_rep, replace=True).mean() for x in range(10000)])
        
        replicate_differences = g1_replicates - g2_replicates
        pvalue = np.sum(replicate_differences >= difference_in_mean) / len(replicate_differences)
        
        mannwhit = stats.mannwhitneyu(g1[genre],g2[genre])
        
        print ((genre, pvalue, mannwhit[1]))
```
